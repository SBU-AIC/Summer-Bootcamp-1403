```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
infections = [2, 5, 18, 28, 43, 61, 95, 139, 245, 388, 593, 978, 1501, 2336, 2922, 3513, 4747, 5823, 6566]
```

```python
x = range(len(infections))
t = np.log(infections)
plt.plot(x, t, 'o' , color='black')
```

```python
def update(x, t, w, b, alpha):
  N = len(x)
  dedw = 0
  dedb = 0
  y = np.dot(w, x) + b
  dedw = np.sum(np.multiply((y - t), x))
  dedb = np.sum(y - t)
  w = w - (alpha/float(N)) * dedw;
  b = b - (alpha/float(N)) * dedb;
  return w, b

def train(x, t, w, b, alpha, epochs):
  for e in range(epochs):
    w, b = update(x, t, w, b, alpha)
    if e % 100 == 0 :
      print("epoch ", e, " loss: ", calc_cost(x, t, w, b))
  return w, b

def calc_cost(x, t, w, b):
  N = len(x)
  err = 0
  y = np.dot(w, x) + b
  err += np.sum(np.multiply((y - t),(y - t)))#(y - t)**2
  return err/float(2*N)
```

```python
w = 10
b = 10
alpha = 0.01
epochs = 100000
w, b = train(x, t, w, b, alpha, epochs)
print(w, b)
```

```python
plt.plot(x, t, 'o', color='red')
yy = np.dot(x, w) + b
plt.plot(x, yy)
```

##Normalization

```python
df_brain_dataset = brain_dataset.copy()
for column in df_brain_dataset.columns:
  df_brain_dataset[column] = (df_brain_dataset[column]-df_brain_dataset[column].min())/(df_brain_dataset[column].max()-df_brain_dataset[column].min())

print(df_brain_dataset)
```

```python
x_df_brain_dataset = df_brain_dataset['Head Size(cm^3)']
y_df_brain_dataset = df_brain_dataset['Brain Weight(grams)']
X = x_df_brain_dataset.to_numpy()
Y = x_df_brain_dataset.to_numpy()
```

```python
w = 10
b = 10
alpha = 0.01
epochs = 100000
w, b = train(X, Y, w, b, alpha, epochs)
print(w, b)
```

```python
plt.plot(X, Y, 'o', color='black')
yy = np.dot(X, w) + b
plt.plot(X, yy)
```

After this we have normalized data and the model can find the line.

##MAE

```python
YPred = np.dot(X, w) + b
N = len(X)
loss = 0
for i in range(N):
  loss = np.abs(Y[i]-YPred[i]) + loss
  #print(loss)
mae = loss/N
print("MAE is: ",mae)
```

##MSE

```python
YPred = np.dot(X, w) + b
N = len(X)
loss = 0
for i in range(N):
  loss = (Y[i]-YPred[i])**2 + loss
mse = loss/N
print("MSE is: ",mse)
```

##RMSE

```python
YPred = np.dot(X, w) + b
N = len(X)
loss = 0
for i in range(N):
  loss = (Y[i]-YPred[i])**2 + loss
mse = loss/N
rmse = np.sqrt(mse)
print("MSE is: ",rmse)
```

##R Squared

```python
YPred = np.dot(X, w) + b
N = len(X)
ssr = 0
ssm = 0
for i in range(N):
  ssr = (Y[i]-YPred[i])**2 + ssr
  ssm = (Y[i]-np.mean(Y))**2 + ssm
r_squared = 1-(ssr/ssm)
print("R Squared error is: ",r_squared)
```
